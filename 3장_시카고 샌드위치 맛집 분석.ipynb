## 반복문으로 랭킹별 정보 추출

```python
import urllib.parse
```

```python
url_base = 'http://www.chicagomag.com'
```

```python
rank =[]
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div','sammy')

for item in list_soup:
    rank.append(item.find(class_='sammyRank').get_text())
    
    tmp_string = item.find(class_='sammyListing').get_text()  # 메뉴이름, 가게이름
    
    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0])  # 메뉴 이름
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])  # 가게 이름
    
    url_add.append(urllib.parse.urljoin(url_base, item.find('a')['href']))   # urllib의 urljoin
```

- urllib의 **urljoin**: 절대경로로 잡힌 url은 그대로 두고, 상대경로로 잡힌 url만 join해서 절대경로로 변경해줌!

```python
main_menu[:5], len(main_menu)
```

```python
cafe_name[:5], len(cafe_name)
```

```python
url_add[:5], len(url_add)
```

```python
import pandas as pd
```

```python
data = {'Rank':rank,'Menu':main_menu,'Cafe':cafe_name,'URL':url_add}
df = pd.DataFrame(data)
df.head()
```

# 50개 웹 페이지에 대한 정보 가져오기

```python
# 상태 진행바 tqdm 적용하고, 50개 url 수집
from tqdm.notebook import tqdm 

price = []
address = []

for n in tqdm(df.index):
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'lxml')
    
    gettings = soup_tmp.find('p','addy').get_text()
    
    price.append(gettings.split()[0][:-1])
    address.append(' '.join(gettings.split()[1:-2]))
```

# 맛집 위치를 지도에 표시 (시각화)

```python
import folium
import googlemaps
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/03. best_sandwiches_list_chicago2.csv', index_col=0)
df.head(5)
```

## 구글맵에서 주소 위도, 경도 가져오기

```python
gmaps_key = '*****'
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
# 주소 Multiple이 나타나지 않는 경우만 주소 검색

lat = []
lng = []

for n in tqdm(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n]+ ', '+'Cicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
# df에 추가
df['lat'] = lat
df['lng'] = lng
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11) # 50개 맛집의 위도, 경도 평균값 구해서 -> 중앙좌표로 사용
folium.Marker([df['lat'].mean(), df['lng'].mean()], popup='center').add_to(mapping)
mapping
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11)
for n in df.index:
    if df['Address'][n] != 'Multiple':
        folium.Marker([df['lat'][n], df['lng'][n]], popup=df['Cafe'][n]).add_to(mapping)
mapping
```

# 네이버 영화 평점 기준으로 영화의 평점 변화 확인하기
- https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date=20171017

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date=20171017'

html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
soup.find_all('div','tit5')  # 1~50위
```

```python
soup.find_all('div','tit5')[0].a.string  # 제목만
```

```python
soup.find_all('td','point')[0].string  # 평점
```

```python
# 20170501~100일간
date = pd.date_range('2017-05-01', periods=100, freq='D')
date
```

```python
import urllib
from tqdm.notebook import tqdm
```

```python
movie_date = []
movie_name = []
movie_point = []

for day in tqdm(date):
    html = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date={date}'
    response = urlopen(html.format(date=urllib.parse.quote(day.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, 'html.parser')
    end = len(soup.find_all('td','point'))
    
    movie_date.extend([day for n in range(0,end)])
    movie_name.extend([soup.find_all('div','tit5')[n].a.string for n in range(0,end)])
    movie_point.extend([soup.find_all('td','point')[n].string for n in range(0,end)])
```

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})
movie.head()
```

```python
movie['point'] = movie['point'].astype(float)
```

```python
# 영화별 점수 합산
movie_best = movie.groupby(['name'])[['point']].sum().sort_values(by='point',ascending=False)
movie_best
```

```python
# . 제거
movie['name'] = movie['name'].map(lambda x: x.replace('.',''))
```

```python
tmp = movie[movie['name'] == '너의 이름은']
tmp
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(tmp['date'], tmp['point'])
plt.grid()
plt.show()
```
