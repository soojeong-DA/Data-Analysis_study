# 대한민국 지도 시각화를 위해 지역별 고유 ID 생성
- 고유 id: 광역시도 + 시도 (단, '경기도 동안구'가 아닌, '안양 동안구'가 되게 수정해야함)

```python
pop['시도'].unique()
```

```python
si_name = [None] * len(pop)
si_name
```

```python
# 광역시가 아니면서, 구를 가지고 있는 시와 해당 행정구 dict
tmp_gu_dict = {
    '수원':['장안구','권선구','팔달구','영통구'],
    '성남':['수정구','중원구','분당구'],
    '안양':['만안구','동안구'],
    '안산':['상록구','단원구'],
    '고양':['덕양구','일산동구','일산서구'],
    '용인':['처인구','기흥구','수지구'],
    '청주':['상당구','서원구','흥덕구','청원구'],
    '천안':['동남구','서북구'],
    '전주':['완산구','덕진구'],
    '포항':['남구','북구'],
    '창원':['의창구','성산구','진해구','마산합포구','마산회원구'],
    '부천':['오정구','원미구','소사구']
}
```

```python
for n in pop.index:
    # 광역시가 아닌 경우
    if pop['광역시도'][n][-3:] not in ['광역시','특별시','자치시']:
        # 고성군 - 강원도, 경상남도 두곳에 존재 => 구분
        if pop['시도'][n][:-1]=='고성' and pop['광역시도'][n]=='강원도':
            si_name[n] = '고성(강원)'
        elif pop['시도'][n][:-1]=='고성' and pop['광역시도'][n]=='경산남도':
            si_name[n] = '고성(경남)'
        # 나머지
        else:
            si_name[n] = pop['시도'][n][:-1]   # 오정구 -> 오정,...
        
        for keys, values in tmp_gu_dict.items():
            if pop['시도'][n] in values:
                if len(pop['시도'][n]) == 2:
                    si_name[n] = keys+ ' ' + pop['시도'][n]
                elif pop['시도'][n] in ['마산합포구','마산회원구']:
                    si_name[n] = keys + ' ' + pop['시도'][n][2:-1]
                else:
                    si_name[n] = keys + ' ' + pop['시도'][n][:-1]
    # 광역시인 경우
    elif pop['광역시도'][n] == '세종특별자치시':
        si_name[n] = '세종'
    else:
        if len(pop['시도'][n])==2:
            si_name[n] = pop['광역시도'][n][:2] + ' ' + pop['시도'][n]
        else:
            si_name[n] = pop['광역시도'][n][:2] + ' ' + pop['시도'][n][:-1]
```

- iterrows(): 코드짤 때 .loc,[i] 등을 붙이지 않아도 되서 편할 수 있으나, 반복문 속도가 느린 것이 단점
    - idx: index
    - row:
        - y      0
        - x      7
        - ID    철원
- annotate: 그래프에 값을 가리키는 화살표와 함께 주석(문자열)을 표시
- plt.gca().invert_yaxis(): y축을 뒤집어줌
    - 원래의 y축에는 큰 숫자가 위에 위치함 -> 뒤집음 => 작은 숫자가 위에 위치함
    - ~.xaxis(): x축을 뒤집어줌
- plt.axis('off'): 그래프의 모든 축과 라벨을 제거해줌 => 그래프 선만 남음
https://m.blog.naver.com/h920203/221684982345

```python
# 경계선 그리기
plt.figure(figsize=(8,11))

# 지역 이름 표시
for idx, row in draw_korea.iterrows():
    
    # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시
    # (중구,서구)
    if len(row['ID'].split()) == 2:
        dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
    elif row['ID'][:2] == '고성':
        dispname = '고성'
    else:
        dispname = row['ID']
    
    # 서대문구, 서귀포시 등 이름이 3자 이상인 경우에 작은 글자로 표시
    if len(dispname.splitlines()[-1]) >= 3:
        fontsize, linespacing = 9.5, 1.5       # 글자 크기, line 스페이스
    else:
        fontsize, linespacing = 11, 1.2
    
    plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold', fontsize=fontsize, ha='center', va='center', linespacing=linespacing)

# 시도 경계 그리기
for path in BORDER_LINES:
    ys, xs = zip(*path)
    plt.plot(xs, ys, c='black', lw=1.5)   # y,x좌표를 연결하는 선 그리기

plt.gca().invert_yaxis()   # y좌표값 중 가장 작은 0이 맨 위에 있어야하기 때문에, y축을 뒤집어줘야함

plt.axis('off')   # 그래프 축, 라벨 등 제거 => 그래프 선만 남게함

plt.tight_layout()   # 공백 적절히 조절해줌(간격 자동 조정..보기좋게!)
plt.show()
```

# 인구 현황 및 인구 소멸지역 확인

```python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75
    
    whitelabelmin = (max(blockedMap[targetData]) - 
                    min(blockedMap[targetData]))*0.25 + min(blockedMap[targetData])
    
    datalabel = targetData
    
    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])
    
    mapdata = blockedMap.pivot_table(index='y',columns='x',values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    # 경계선 그리기
    plt.figure(figsize=(9,11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, edgecolor='#aaaaaa', linewidth=0.5)

    
    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():

        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시
        # (중구,서구)
        if len(row['ID'].split()) == 2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2] == '고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        # 서대문구, 서귀포시 등 이름이 3자 이상인 경우에 작은 글자로 표시
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1       # 글자 크기, line 스페이스
        else:
            fontsize, linespacing = 11, 1.
        
        annocolor ='white' if row[targetData] > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold', fontsize=fontsize, ha='center', va='center', 
                     color=annocolor, linespacing=linespacing)

    # 시도 경계 그리기
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)   # y,x좌표를 연결하는 선 그리기

    plt.gca().invert_yaxis()   # y좌표값 중 가장 작은 0이 맨 위에 있어야하기 때문에, y축을 뒤집어줘야함

    plt.axis('off')   # 그래프 축, 라벨 등 제거 => 그래프 선만 남게함
    
    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)
    
    plt.tight_layout()   # 공백 적절히 조절해줌(간격 자동 조정..보기좋게!)
    plt.show()
```

```python
# 만든 drawKorea를 사용해, 인구수합계 그려보기
drawKorea('인구수합계', pop, 'Blues')
```

# 인구 현황에서 여성 인구 비율 확인

- 데이터에 음(-)의 값이 있는지 여부에 따라 일부 설정 변경 필요

```python
def drawKorea2(targetData, blockedMap, cmapname):
    gamma = 0.75
    
    whitelabelmin = 20.
    
    datalabel = targetData
    
    tmp_max = max([np.abs(min(blockedMap[targetData])),
                      np.abs(max(blockedMap[targetData]))]) # 절대값으로
    
    vmin, vmax = -tmp_max, tmp_max
    
    mapdata = blockedMap.pivot_table(index='y',columns='x',values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    # 경계선 그리기
    plt.figure(figsize=(9,11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, edgecolor='#aaaaaa', linewidth=0.5)

    
    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():

        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시
        # (중구,서구)
        if len(row['ID'].split()) == 2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2] == '고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        # 서대문구, 서귀포시 등 이름이 3자 이상인 경우에 작은 글자로 표시
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1       # 글자 크기, line 스페이스
        else:
            fontsize, linespacing = 11, 1.
        
        annocolor ='white' if np.abs(row[targetData]) > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold', fontsize=fontsize, ha='center', va='center', 
                     color=annocolor, linespacing=linespacing)

    # 시도 경계 그리기
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)   # y,x좌표를 연결하는 선 그리기

    plt.gca().invert_yaxis()   # y좌표값 중 가장 작은 0이 맨 위에 있어야하기 때문에, y축을 뒤집어줘야함

    plt.axis('off')   # 그래프 축, 라벨 등 제거 => 그래프 선만 남게함
    
    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)
    
    plt.tight_layout()
    plt.show()
```

```python
pop['여성비'] = (pop['인구수여자']/pop['인구수합계'] - 0.5)*100    # -0.5 => 0 이상 값이면 여성 인구수가 50% 이상인 것 (양수,음수 기준으로 색상 달리 하려고했음)
drawKorea2('여성비', pop, 'RdBu')
```

- 모든 연령의 여성 인구 비중을 구한 것이라, 여성 비중이 높은 곳이 많이 눈에 띔
    - 특히, 서울, 부산, 그리고 고연령층이 많이 분포하는 전라남도 북도 등

```python
# 2030세대 기준으로 지도 시각화
pop['2030여성비'] = (pop['20-39세여자']/pop['20-39세합계'] - 0.5)*100
drawKorea2('2030여성비', pop, 'RdBu')
```

- 고양시, 서울 강남부근, 세종 등의 일부 지역을 제외한 나머지에서 남성 비중이 더 높게 나타남
- 고령 여성 인구를 제외한 2030세대는 전국적으로 남성 비율이 더 높다는 것을 알 수 있음
