```python
import googlemaps
```

```python
gmaps_key = "****"
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
gmaps.geocode('서울중부경찰서', language='ko')
#formatted_address - 주소
```

```python
# 구글 검색에 맞게  중부서,수서서 등 -> 서울**경찰서로 변환해주기
# 중부'서' -> '서'빼주기[:-1]
station_name = []
for name in crime_anal_police['관서명']:
    station_name.append('서울'+str(name[:-1])+'경찰서')

station_name
```

```python
# google map에서 경찰서명 검색해서 주소,위도,경도 가져오기
station_address = []
station_lat = []
station_lng = []

for name in station_name:
    tmp = gmaps.geocode(name, language = 'ko')   # 경찰서 이름 - 검색
    station_address.append(tmp[0].get('formatted_address'))  #주소 -> station_address
    
    tmp_loc = tmp[0].get('geometry') #위도,경도
    station_lat.append(tmp_loc['location']['lat'])  # location(key) dict에서 lat(value) 가져오기
    station_lng.append(tmp_loc['location']['lng'])
    print(name + '-->' + tmp[0].get('formatted_address'))
```

# [Tutorial] pandas pivot_table 학습
- pivot_table 사용해서 합치기
    - name 컬럼이 index가됨
    - 중복 name 컬럼 값이 합쳐지면서, values들은 기본적으로 평균값(np.mean)을 취하게됨
    - aggfunc 옵션을 사용하면, sum도 가능(np.sum)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('02. sales-funnel.xlsx')
df.head()
```

```python
# pivot table 사용
# 중복 name의 value는 기본적으로 평균을 취함
# 기본적으로 숫자형 data 컬럼만 나옴
pd.pivot_table(df, index = ['Name'])
```

```python
# index 여러개 지정
pd.pivot_table(df, index= ['Name', 'Rep','Manager'])
```

```python
# 특정 value만 지정
pd.pivot_table(df, index = ['Manager', 'Rep'], values = ['Price'])
```

```python
# 평균말고 합계로 옵션 변경.
pd.pivot_table(df, index = ['Manager', 'Rep'], values = ['Price'], aggfunc = np.sum)
```

```python
# 해당 과정에서 빈칸이 나타나면, NaN 표시 변경 옵션: fill_value
pd.pivot_table(df, index = ['Manager', 'Rep','Product'],
              values = ['Price', 'Quantity'],
              aggfunc = [np.sum, np.mean],
              fill_value = 0,      # 빈값은 0으로 채워라
              margins = True)      # 총계(ALL)를 내줌
```

# pivot_table 이용해 데이터 정리

```python
# 아까 저장했던 파일 불러오기
crime_anal_raw = pd.read_csv('02. crime_in_seoul_include_gu_name.csv',
                            encoding = 'utf-8')
crime_anal_raw.head()
```

```python
# pivot_table 이용해서, 관서별 -> 구별 바꾸기  (관서가 중요한게 아니라, '구'가 중요)
crime_anal_rqw = pd.read_csv('02. crime_in_seoul_include_gu_name.csv',
                            encoding = 'utf-8', index_col = 0)
crime_anal = pd.pivot_table(crime_anal_raw, index = '구별', aggfunc = np.sum)
crime_anal.head()
```

```python
# 각 범죄별 검거율 계산 : 검거 / 발생 * 100
crime_anal['강간검거율'] = crime_anal['강간 검거'] / crime_anal['강간 발생'] * 100
crime_anal['강도검거율'] = crime_anal['강도 검거'] / crime_anal['강도 발생'] * 100
crime_anal['살인검거율'] = crime_anal['살인 검거'] / crime_anal['살인 발생'] * 100
crime_anal['절도검거율'] = crime_anal['절도 검거'] / crime_anal['절도 발생'] * 100
crime_anal['폭력검거율'] = crime_anal['폭력 검거'] / crime_anal['폭력 발생'] * 100

# 검거 건수는 검거율로 대체가능 -> 삭제 처리
del crime_anal['강간 검거']
del crime_anal['강도 검거']
del crime_anal['살인 검거']
del crime_anal['절도 검거']
del crime_anal['폭력 검거']

crime_anal.head()
```

```python
# unnamed 컬럼 제거
crime_anal.drop(crime_anal.columns[0:1], axis = 1, inplace = True)
crime_anal.head()
```

```python
# 검거율에 100이 넘는 값들이 있음 -> 아마, 전년도 발생 건수에 대한 검거도 포함되서 그런듯
# 지금 하는건 학습 목적이니, 살짝 넘어가고, 100으로만 처리해주자ㅠㅠ!
con_list = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']

# 각 컬럼별로 100넘는 값을 100으로 변경해줌  # > 100 : True/False로 반환됨
for column in con_list:
    crime_anal.loc[crime_anal[column] > 100, column] = 100  
    
crime_anal.head()    
```

```python
# '발생' 단어 제거  - 컬럼 이름 수정: rename
crime_anal.rename(columns = {'강간 발생':'강간',
                            '강도 발생': '강도',
                            '살인 발생':'살인',
                            '절도 발생':'절도',
                            '폭력 발생':'폭력'}, inplace = True)

crime_anal.head()
```

# 데이터 시각화를 위해 데이터 다듬기
- 강도, 살인은 두자릿수 범위 but, 절도, 폭력은 네자릿수  => **컬럼별 수치 차이가 큼**
- **컬럼별로 `정규화` 시키기**: 각 컬럼의 **최대값을 1**로 설정하고, 이에 맞춰 나머지 각 항목 값 변경

- scikit learn(sklearn) 모듈
    - 전처리(preprocessing) 도구에는 최소값, 최대값을 이용해 정규화 시키는 함수 존재
    - .MinMaxScaler(), .fit_transform()

```python
# 각 컬럼의 최대값에 맞춰서, 정규화 시키기
from sklearn import preprocessing

col = ['강간','강도','살인','절도','폭력']  #정규화 대상 컬럼 

x = crime_anal[col].values   # 각 컬럼 값을 담은 row가 한줄씩 리스트에 담아짐(array)
min_max_scaler = preprocessing.MinMaxScaler()

x_scaled = min_max_scaler.fit_transform(x.astype(float))
crime_anal_norm = pd.DataFrame(x_scaled, columns = col, index = crime_anal.index)

col2 = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crime_anal_norm[col2] = crime_anal[col2]
crime_anal_norm.head()
```

## 1장 실습 data에서  인구수, CCTV 개수 가지고오기

```python
# file 불러오기
result_cctv = pd.read_csv('01. CCTV_result.csv', encoding = 'UTF-8', index_col = '구별')

# 인구수, cctv개수 컬럼 추가, data 대입
crime_anal_norm[['인구수','CCTV']] = result_cctv[['인구수', '소계']]
crime_anal_norm.head()
```

```python
# '범죄'라는 컬럼을 생성해, 각 범죄 발생건수 합 대입 (통합)
col = ['강간','강도','살인','절도','폭력']
crime_anal_norm['범죄'] = np.sum(crime_anal_norm[col], axis =1)
crime_anal_norm.head()
```

```python
# '검거' 컬럼 생성 및 검거율 통합
col = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crime_anal_norm['검거'] = np.sum(crime_anal_norm[col], axis = 1)
crime_anal_norm.head()
```

- 검거율 합계인 **'검거' 컬럼의 최대값을 100**으로 하는 값 변환을 진행(각 행의 '검거' 값을 '검거'최대값으로 나누기)

- 서부쪽은 범죄도 많이 발생하지만, 검거율도 높다
- 강북쪽은 범죄가 많이 발생하지않고, 검거율도 높징 않은편
